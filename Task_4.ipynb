```
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')


def preprocess_text(text):
    tokens=word_tokenize(text.lower())

    # Remove punctuation
    table=str.maketrans('', '',string.punctuation)
    tokens=[w.translate(table) for w in tokens]

    # Remove remaining tokens that are not alphabetic
    tokens=[word for word in tokens if word.isalpha()]

    # Remove stopwords
    stop_word=set(stopwords.words('english'))
    tokens=[word for word in tokens if not word in stop_words]

    # Lemmatization
    lemmatizer=WordNetLemmatizer()
    tokens=[lemmatizer.lemmatize(word) for word in tokens]


    preprocessed_text=' '.join(tokens)

    return preprocessed_text


```

```
train_data=train_data.assign(preprocessed_text=train_data['text'].apply(preprocess_text))
train_data
```

```
dev_seen=dev_seen.assign(preprocessed_text=dev_seen['text'].apply(preprocess_text))
dev_unseen=dev_unseen.assign(preprocessed_text=dev_unseen['text'].apply(preprocess_text))
test_seen=test_seen.assign(preprocessed_text=test_seen['text'].apply(preprocess_text))
test_unseen=test_unseen.assign(preprocessed_text=test_unseen['text'].apply(preprocess_text))
```

##HERE I'M PERFORMING SENTIMENT ANALYSIS ON THE TEXT
##IM ASSINGING THE FOLLOWING NUMERIC VALUES FOR THE FOLLOWING-
##1.POSITIVE:+1
##2.NEUTRAL:0
##3.NEGATIVE:-1

```
from textblob import TextBlob
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity < 0:
        return -1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return 1
```

##APPLYING THE VALUES HERE

```
train_data['sentiment_analysis']=train_data['preprocessed_text'].apply(analyze_sentiment)
```

```
test_seen['sentiment_analysis']=test_seen['preprocessed_text'].apply(analyze_sentiment)
test_unseen['sentiment_analysis']=test_unseen['preprocessed_text'].apply(analyze_sentiment)
dev_seen['sentiment_analysis']=dev_seen['preprocessed_text'].apply(analyze_sentiment)
dev_unseen['sentiment_analysis']=dev_unseen['preprocessed_text'].apply(analyze_sentiment)
```

#PERFORMING UNIMODAL ANALYSIS

##LOGISTIC REGRESSION

##I AM USING TF-IDF FOR VECTORIZATION OF FEATURE EXTRACTION AND DIMENSIONALITY REDUCTION

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np


def train_and_evaluate_model(balanced_train_data,test_unseen_data,test_seen_data,dev_seen_data,dev_unseen_data):

  tfidf_vectorizer=TfidfVectorizer(max_features=10)
  X_train_text=tfidf_vectorizer.fit_transform(balanced_train_data['preprocessed_text'])

  # Combining text and sentiment features for training data
  X_train=np.hstack((X_train_text.toarray(),balanced_train_data['sentiment_analysis'].values.reshape(-1, 1)))

  logistic_regression_model=LogisticRegression(C=1.0, random_state=42)
  logistic_regression_model.fit(X_train, balanced_train_data['label'])


  def evaluate_data(data, name):
    X_data_text=tfidf_vectorizer.transform(data['preprocessed_text'])
    X_data=np.hstack((X_data_text.toarray(), data['sentiment_analysis'].values.reshape(-1, 1)))
    predictions=logistic_regression_model.predict(X_data)
    accuracy=accuracy_score(data['label'],predictions)
    print(f"{name} Accuracy:",accuracy)

  evaluate_data(test_unseen_data,"Test Unseen")
  evaluate_data(test_seen_data,"Test seen")
  evaluate_data(dev_seen_data,"Dev seen")
  evaluate_data(dev_unseen_data,"Dev Unseen")


train_and_evaluate_model(balanced_train_data,test_unseen,test_seen,dev_seen,dev_unseen)

```

#DECISION TREES


```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np

def train_and_evaluate_model(balanced_train_data,test_unseen_data,test_seen_data,dev_seen_data,dev_unseen_data):
    tfidf_vectorizer=TfidfVectorizer(max_features=10)
    X_train_text=tfidf_vectorizer.fit_transform(balanced_train_data['preprocessed_text'])
    X_train=np.hstack((X_train_text.toarray(),balanced_train_data['sentiment_analysis'].values.reshape(-1, 1)))

    #hyperparameters
    hyperparameters={
        'max_depth':20,
        'min_samples_split':2,
        'min_samples_leaf':2,
        'max_features':None
    }

    decision_tree_model=DecisionTreeClassifier(**hyperparameters)
    decision_tree_model.fit(X_train, balanced_train_data['label'])


    def evaluate_data(data, name):
        X_data_text=tfidf_vectorizer.transform(data['preprocessed_text'])
        X_data=np.hstack((X_data_text.toarray(),data['sentiment_analysis'].values.reshape(-1, 1)))
        predictions=decision_tree_model.predict(X_data)
        accuracy=accuracy_score(data['label'],predictions)
        print(f"{name} Accuracy:",accuracy)

    evaluate_data(test_unseen_data,"Test Unseen")
    evaluate_data(test_seen_data,"Test seen")
    evaluate_data(dev_seen_data,"Dev seen")
    evaluate_data(dev_unseen_data,"Dev Unseen")


train_and_evaluate_model(balanced_train_data,test_unseen,test_seen,dev_seen,test_unseen)


```

#RANDOM FOREST

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

def train_and_evaluate_model(balanced_train_data,test_unseen_data,test_seen_data,dev_seen_data,dev_unseen_data):

    tfidf_vectorizer=TfidfVectorizer(max_features=10)
    X_train_text=tfidf_vectorizer.fit_transform(balanced_train_data['preprocessed_text'])


    X_train=np.hstack((X_train_text.toarray(),balanced_train_data['sentiment_analysis'].values.reshape(-1, 1)))

    #hyperparameters
    hyperparameters={
        'n_estimators':100,
        'max_depth':20,
        'min_samples_split':2,
        'min_samples_leaf':2,
        'max_features':'auto'
    }

    random_forest_model=RandomForestClassifier(**hyperparameters)
    random_forest_model.fit(X_train balanced_train_data['label'])

    def evaluate_data(data, name):
        X_data_text=tfidf_vectorizer.transform(data['preprocessed_text'])
        X_data=np.hstack((X_data_text.toarray(), data['sentiment_analysis'].values.reshape(-1, 1)))
        predictions=random_forest_model.predict(X_data)
        accuracy=accuracy_score(data['label'],predictions)
        print(f"{name} Accuracy:",accuracy)

    evaluate_data(test_unseen_data,"Test Unseen")
    evaluate_data(test_seen_data,"Test Seen")
    evaluate_data(dev_seen_data,"Dev Seen")
    evaluate_data(dev_unseen_data,"Dev Unseen")


train_and_evaluate_model(balanced_train_data,test_unseen,test_seen,dev_seen,dev_unseen)
```
